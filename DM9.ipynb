```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from IPython.display import display

# 1. Load dataset
df = pd.read_csv('transaction.csv')
print("1. Dataset (5 baris pertama):")
display(df.head())

# 2. Hitung kemunculan tiap negara
country_counts = df['Country'].value_counts().reset_index()
country_counts.columns = ['Country', 'Jumlah Kemunculan']
print("2. Kemunculan tiap negara:")
display(country_counts)

# 3. Hitung rata-rata jumlah barang (Qty) per transaksi per negara
# 1 transaksi = 1 InvoiceNo
qty_per_transaksi = df.groupby(['Country', 'InvoiceNo'])['Qty'].sum().reset_index()
avg_qty_per_country = qty_per_transaksi.groupby('Country')['Qty'].mean().reset_index()
avg_qty_per_country.columns = ['Country', 'RataRataQty']
print("3. Rata-rata Qty per transaksi per negara:")
display(avg_qty_per_country)

# 4. Clustering 10 kali dengan KMeans (k=3), simpan SSE untuk tiap clustering
best_sse = float('inf')
best_model = None
best_labels = None

X = avg_qty_per_country[['RataRataQty']].values
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

for i in range(10):
    kmeans = KMeans(n_clusters=3, random_state=i)
    kmeans.fit(X_scaled)
    sse = kmeans.inertia_
    if sse < best_sse:
        best_sse = sse
        best_model = kmeans
        best_labels = kmeans.labels_

# 5. Ambil cluster terbaik
avg_qty_per_country['Cluster'] = best_labels
print("5. Cluster hasil terbaik berdasarkan SSE:")
display(avg_qty_per_country)

# 6. Tentukan posisi centroid
centroids = best_model.cluster_centers_.flatten()
centroids_unscaled = scaler.inverse_transform(best_model.cluster_centers_).flatten()
centroid_df = pd.DataFrame({
    'Cluster': range(3),
    'Centroid': centroids_unscaled
})
print("6. Centroid tiap cluster:")
display(centroid_df)

# 7. Urutkan centroid secara ascending
sorted_centroids = centroid_df.sort_values(by='Centroid').reset_index(drop=True)
print("7. Sorted centroid:")
display(sorted_centroids)

# 8. Label cluster berdasarkan posisi centroid
cluster_map = {}
sorted_cluster_ids = sorted_centroids['Cluster'].tolist()
cluster_map[sorted_cluster_ids[0]] = 'Rendah'
cluster_map[sorted_cluster_ids[1]] = 'Sedang'
cluster_map[sorted_cluster_ids[2]] = 'Tinggi'
avg_qty_per_country['Label Cluster'] = avg_qty_per_country['Cluster'].map(cluster_map)

print("8. Negara dengan label cluster (rendah, sedang, tinggi):")
display(avg_qty_per_country[['Country', 'RataRataQty', 'Label Cluster']])

# 9. Visualisasi
plt.figure(figsize=(12, 6))
color_map = {'Rendah': 'red', 'Sedang': 'orange', 'Tinggi': 'green'}

# Urutkan berdasarkan nama negara agar sumbu X konsisten
sorted_data = avg_qty_per_country.sort_values('Country').reset_index(drop=True)

plt.bar(x=range(len(sorted_data)), 
        height=sorted_data['RataRataQty'], 
        color=sorted_data['Label Cluster'].map(color_map))

plt.xticks(ticks=range(len(sorted_data)), labels=sorted_data['Country'], rotation=90)
plt.ylabel('Rata-rata Qty per Transaksi')
plt.title('9. Cluster Rata-rata Qty per Transaksi per Negara')
plt.tight_layout()
plt.show()
```
