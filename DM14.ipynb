```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, fcluster
from sklearn.linear_model import LinearRegression
from IPython.display import display

# 1. Import dan tampilkan dataset
print("1. Load dataset:")
df = pd.read_csv("transaction.csv")
display(df.head())

# 2. Ambil fitur penting dari tahun 2011
print("2. Filter fitur: InvoiceNo, Country, Month dari tahun 2011")
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df['year'] = df['InvoiceDate'].dt.year
df['month'] = df['InvoiceDate'].dt.month

data = df[df['year'] == 2011][['InvoiceNo', 'Country', 'month']]
display(data.head())

# 3–8. Iterasi per bulan dan clustering
print("3–8. Iterasi bulan 1–12: clustering dan pengambilan transaksi tinggi per bulan")

transaksi_tinggi = []
rekap_bulanan = []

for i in range(1, 13):
    print(f"\n--- Bulan ke-{i} ---")

    # Ambil data bulan ke-i
    data_bulan = data[data['month'] == i]

    # Hitung banyaknya transaksi per negara
    transaksi = data_bulan.groupby('Country')['InvoiceNo'].nunique().reset_index()
    transaksi.columns = ['Country', 'Transaksi']

    # Clustering dengan Average Linkage
    X = transaksi[['Transaksi']].values
    linkage_matrix = linkage(X, method='average')
    clusters = fcluster(linkage_matrix, t=3, criterion='maxclust')
    transaksi['Cluster'] = clusters

    # Hitung centroid tiap cluster
    centroids = transaksi.groupby('Cluster')['Transaksi'].mean().reset_index()
    centroids.columns = ['Cluster', 'Centroid']

    # Urutkan centroid
    sorted_centroids = centroids.sort_values(by='Centroid').reset_index(drop=True)

    # Ambil nilai cluster tertinggi
    cluster_tertinggi_id = sorted_centroids.iloc[-1]['Cluster']
    transaksi_tertinggi = sorted_centroids.iloc[-1]['Centroid']
    transaksi_tinggi.append(transaksi_tertinggi)

    # Simpan rekap
    rekap_bulanan.append({
        'Bulan': i,
        'Cluster Tertinggi': cluster_tertinggi_id,
        'Transaksi Tinggi': transaksi_tertinggi
    })

    # Output tabel dan detail bulan ini
    print("Transaksi per negara (Top 5):")
    display(transaksi[['Country', 'Transaksi', 'Cluster']].sort_values(by='Transaksi', ascending=False).head())

    print("Centroid per cluster:")
    display(centroids)

    print(f"Cluster tertinggi: {cluster_tertinggi_id} | Nilai transaksi tinggi bulan {i}: {transaksi_tertinggi:.2f}")

# 9. Visualisasi pergerakan transaksi tinggi
print("\n9. Visualisasi pergerakan nilai transaksi tinggi:")

plt.figure(figsize=(10, 6))
plt.plot(range(1, 13), transaksi_tinggi, marker='o', color='blue')
plt.title('Pergerakan Nilai Transaksi Tinggi per Bulan (2011)')
plt.xlabel('Bulan')
plt.ylabel('Jumlah Transaksi Tinggi')
plt.grid(True)
plt.xticks(range(1, 13))
plt.tight_layout()
plt.show()

rekap_df = pd.DataFrame(rekap_bulanan)
print("\nTabel Rekap Transaksi Tinggi per Bulan:")
display(rekap_df)

# 10. Prediksi Januari 2012 dengan Linear Regression
print("\n10. Prediksi nilai transaksi tinggi bulan Januari 2012:")

X = np.array(range(1, 13)).reshape(-1, 1)
y = np.array(transaksi_tinggi).reshape(-1, 1)

model = LinearRegression()
model.fit(X, y)
predicted_val = model.predict(np.array([[13]]))  # Prediksi bulan ke-13

print(f"Prediksi transaksi tinggi Januari 2012: {predicted_val[0][0]:.2f}")
```
